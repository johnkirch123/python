```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apple_data = pd.read_csv('AAPL_pyplot.csv')
thirty_day_apple_data = apple_data[0:30]
thirty_day_apple_data
```

```python
dates = thirty_day_apple_data['Date']
opens = thirty_day_apple_data['Open']
closes = thirty_day_apple_data['Close']
highs = thirty_day_apple_data['High']
lows = thirty_day_apple_data['Low']
```

```python
plt.plot(dates, closes)
plt.plot(dates, opens)
plt.show()
```

```python
plt.plot(dates, highs)
plt.plot(dates, lows)
plt.show()
```

```python
fig = plt.figure()
plt.subplot(221)
plt.plot(dates, opens)
plt.subplot(222)
plt.plot(dates, closes)
plt.subplot(223)
plt.plot(dates, highs)
plt.subplot(224)
plt.plot(dates, lows)
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(dates, opens)
axes[0, 1].plot(dates, closes)
axes[1, 0].plot(dates, highs)
axes[1, 1].plot(dates, lows)
plt.show()
```

```python
plt.plot(dates, closes)
plt.title('Close Prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')
plt.xticks(pd.date_range('2019-03-18', '2019-04-25', 6).strftime('%Y-%m-%d'))
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
fig.suptitle('Prices for AAPL', y=1.05)
axes[0].plot(dates, highs)
axes[0].set_title('High prices for AAPL')
axes[0].set_xlabel('Dates')
axes[0].set_ylabel('Prices in USD')
axes[1].plot(dates, lows)
axes[1].set_title('Low prices for AAPL')
axes[1].set_xlabel('Dates')
axes[1].set_ylabel('Prices in USD')
plt.tight_layout()
plt.show()
```

```python
plt.plot(dates, highs)
plt.plot(dates, lows)
plt.title('Daily High/Low Prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')
plt.xticks(pd.date_range('2019-03-18', '2019-04-25', 6).strftime('%Y-%m-%d'))
plt.legend(('High', 'Low'), loc='upper left')
plt.show()
```

```python
plt.plot(dates, closes, 'g>-.', lw=3)
plt.title('Close Prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')
plt.xticks(pd.date_range('2019-03-18', '2019-04-25', 6).strftime('%Y-%m-%d'))
plt.show()
```

```python
plt.scatter(dates, closes)
plt.show()
```

```python
plt.stackplot(dates, (highs, lows))
```

```python
plt.hist(highs, bins=len(highs))
plt.show()
```
